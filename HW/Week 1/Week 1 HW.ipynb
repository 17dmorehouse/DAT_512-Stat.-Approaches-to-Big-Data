d.) Run the example shown in section 4.3 of McKinney, of the image plot of the NumPy array of square root of x^2+y^2.

```python
points = np.arange(-5, 5, 0.01) 
xs, ys = np.meshgrid(points, points)
z = np.sqrt(xs ** 2 + ys ** 2)
plt.imshow(z, cmap=plt.cm.gray); plt.colorbar()
plt.title("Image plot of $\sqrt{x^2 + y^2}$ for a grid of values")
```

e.) This example is from the NumPy site at
https://numpy.org/doc/stable/user/absolute_beginners.html
use the plot_surface() function  to show your results in d.   Look up contour plots on matplotlib and show the data from d as a contour plot.

```python
fig = plt.figure()

ax = fig.add_subplot(projection='3d')

points = np.arange(-5, 5, 0.01) 
xs, ys = np.meshgrid(points, points)

R = np.sqrt(xs**2 + ys**2)

Z = np.sin(R)


ax.plot_surface(xs, ys, Z, rstride=1, cstride=1, cmap='viridis')
```
