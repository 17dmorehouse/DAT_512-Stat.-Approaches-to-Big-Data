**The MPG data set and setting up ggplot graphics**

*Data* is the information to use when creating the plot.

*Aesthetics* (aes) provides a mapping between data variables and aesthetic, or graphical, variables used by the underlying drawing system. In the previous section, you mapped the date and pop data variables to the x- and y-axis aesthetic variables.

*Geometric objects* (geoms) defines the type of geometric object to use in the drawing. You can use points, lines, bars, and many others.

**Optional GGplot components**

**Statistical transformations** specify computations and aggregations to be applied to the data before plotting it.

**Scales** apply some transformation during the mapping from data to aesthetics. For example, sometimes you can use a logarithmic scale to better reflect some aspects of your data.

**Facets** allow you to divide data into groups based on some attributes and then plot each group into a separate panel in the same graphic.

**Coordinates systems** map the position of objects to a 2D graphical location in the plot. For example, you can choose to flip the vertical and horizontal axes if that makes more sense in the visualization youâ€™re building.

**Themes** allows you to control visual properties like colors, fonts, and shapes.


````python
# here is the mpg data set

from plotnine.data import mpg

# set up the plot and define the aesthetics, note that the axes are set, the variables are show but without a geom, nothing plots

ggplot(mpg) + aes(x="class", y="hwy")
````

````python
# here is a scatterplot using this data

from plotnine import geom_point


ggplot(mpg) + aes(x="class", y="hwy") + geom_point()
````

````python
#here is a barplot version
# note the need to explicitly import each geom

from plotnine import geom_bar

ggplot(mpg) + aes(x="class") + geom_bar()
````

**Aggregate and Transform**

````python
# Import our example dataset with the levels of Lake Huron 1875â€“1975
from plotnine.data import huron

huron
````

**Group the level measurements into bins**, show the counts per bin (ie basic Histogram)

````python

from plotnine import stat_bin
ggplot(huron) + aes(x="level") + stat_bin(bins=10) + geom_bar()
````

````python
#Here is an actual histogram

from plotnine import  geom_histogram

ggplot(huron) + aes(x="level") + geom_histogram(bins=10)
````

**Group Data  by Decade**

````python
from plotnine import geom_boxplot

(
  ggplot(huron)
  + aes(x="factor(decade)", y="level")
  + geom_boxplot()
)
````

**Coordinate Flip**

````python
from plotnine import coord_flip

ggplot(mpg) + aes(x="class") + geom_bar() + coord_flip()
````

**Faceting**

Group data by some attribute and the plot each group separately

````python
from plotnine import  facet_grid, labs

# note how labs() is used to set labels in this implementation

(
    ggplot(mpg)
    + facet_grid(facets="year~class")
    + aes(x="displ", y="hwy")
    + labs(
        x="Engine Size",
        y="Miles per Gallon",
        title="Miles per Gallon for Each Year and Vehicle Class",
    )
    + geom_point()
)
````

**Themes**

Remember how Themes could be used to control the format details in ggplot?

````python
from plotnine import theme_dark

(
    ggplot(mpg)
    + facet_grid(facets="year~class")
    + aes(x="displ", y="hwy")
    + labs(
        x="Engine Size",
        y="Miles per Gallon",
        title="Miles per Gallon for Each Year and Vehicle Class",
    )
    + geom_point()
    + theme_dark()
)
````

````python
from plotnine import theme_xkcd

(
    ggplot(mpg)
    + facet_grid(facets="year~class")
    + aes(x="displ", y="hwy")
    + labs(
        x="Engine Size",
        y="Miles per Gallon",
        title="Miles per Gallon for Each Year and Vehicle Class",
    )
    + geom_point()
    + theme_xkcd()
)
````

**Exporting a plot image to a file**

````python
# save the plot to a varialbe

myPlot=(
    ggplot(mpg)
    + facet_grid(facets="year~class")
    + aes(x="displ", y="hwy")
    + labs(
        x="Engine Size",
        y="Miles per Gallon",
        title="Miles per Gallon for Each Year and Vehicle Class",
    )
    + geom_point()
    + theme_xkcd()
)

#Call the member function write to a file

myPlot.save("myplot.png",dpi=600)

#note, if you run this code on Google Colab, it will be save to the local temporary file system on colab
# you can download it from there
````
