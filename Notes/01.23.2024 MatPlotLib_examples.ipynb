Matplotlib is pretty close to being a "port" of the plot routines in matlab

Many other more sophisticated plotting programs in Python are based on Matplotlib routines

Examples taken from

https://matplotlib.org/stable/plot_types/index.html

```python
# imports

import matplotlib.pyplot as plt
import numpy as np

# there is a plot style function,  akin to a theme
# you don't have to use them,  try running this with the plt.style command commented out

plt.style.use('_mpl-gallery')

# Numpy creating an array from o to 100 by 10,   and then y is  a sin function
# based on x

# make data
x = np.linspace(0, 10, 100)
y = 4 + 2 * np.sin(2 * x)

# basic plot operation, settting up a figure and an axis

# plot
fig, ax = plt.subplots()

#plot the two lines

ax.plot(x, y, linewidth=2.0)

#control the axis limits

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

#show the plot

plt.show()
```

```python
# really simple, fast version
# this looks like a default plot in matlab

plt.plot(x,y)
```

```python
# here is a scatter plot instead of a lineplot

plt.scatter(x,y)
```

```python
# Here is a barplot

plt.style.use('_mpl-gallery')

# make data:
x = 0.5 + np.arange(8)
y = [4.8, 5.5, 3.5, 4.6, 6.5, 6.6, 2.6, 3.0]

# plot
fig, ax = plt.subplots()

ax.bar(x, y, width=1, edgecolor="white", linewidth=0.7)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

```python
# Simple histogram

# make data
np.random.seed(1)
x = 4 + np.random.normal(0, 1.5, 200)

plt.hist(x, bins=8, linewidth=0.5, edgecolor="white")
```

```python
#fancier histogram

plt.style.use('_mpl-gallery')

# make data
np.random.seed(1)
x = 4 + np.random.normal(0, 1.5, 200)

# plot:
fig, ax = plt.subplots()

ax.hist(x, bins=8, linewidth=0.5, edgecolor="white")

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 56), yticks=np.linspace(0, 56, 9))

plt.show()
```

```python
#Fairly ornate boxplot

plt.style.use('_mpl-gallery')

# make data:
np.random.seed(10)
D = np.random.normal((3, 5, 4), (1.25, 1.00, 1.25), (100, 3))

# plot
fig, ax = plt.subplots()
VP = ax.boxplot(D, positions=[2, 4, 6], widths=1.5, patch_artist=True,
                showmeans=False, showfliers=False,
                medianprops={"color": "white", "linewidth": 0.5},
                boxprops={"facecolor": "C0", "edgecolor": "white",
                          "linewidth": 0.5},
                whiskerprops={"color": "C0", "linewidth": 1.5},
                capprops={"color": "C0", "linewidth": 1.5})

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```
